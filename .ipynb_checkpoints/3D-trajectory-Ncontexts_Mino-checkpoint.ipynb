```python
# This code below is to set the structure of the code for Pyramidal Neuron - Larkum style
# Learning through target spikes

import numpy as np
import matplotlib.pyplot as plt
from importlib import reload
from tqdm import trange
import random
import pandas as pd
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

import lttb_contesto_both as lttb_module
#from lttb_contesto_sopra import LTTB

N, I, O, T = 500, 5, 3, 1000
shape = (N, I, O, T)

dt = .001
tau_m = 20. * dt
tau_s = 2. * dt
tau_ro = 10. * dt
tau_star = 20. * dt
tau_W = 200 * dt


beta = np.exp(-dt/tau_s)
beta_ro = np.exp(-dt/tau_ro)
beta_targ = np.exp(-dt/tau_star)
beta_W = np.exp(-dt/tau_W)

# std of J's used to project context - it used to be 20
sigma_apical_context = 0
sigma_basal_context = 0

sigma_targ = 5. # 10**1.5
sigma_in = 12.

dv = 1 / 500.
alpha = .005
alpha_rout = .01
Vo = - 4
h = - 1
s_inh = 20

sigma_Jrec = 0.
sigma_wrec = 0.
sigma_Jout = 0.1

Ne = 400
Ni = 100

N = Ne+Ni

n_contexts = 2

# Here we build the dictionary of the simulation parameters
par = {'dt' : dt, 'tau_m' : tau_m, 'tau_s' : tau_s, 'tau_ro' : tau_ro, 'tau_star' : tau_star,'tau_W' : tau_W,
       'dv' : dv, 'Vo' : Vo, 'h' : h, 's_inh' : s_inh,
       'N' : N, 'Ni' : Ni, 'Ne' : Ne, 'T' : T, 'I' : I, 'O' : O, 'shape' : shape, 
       'sigma_Jrec' : sigma_Jrec,'sigma_wrec' : sigma_wrec, 'sigma_Jout' : sigma_Jout,  'n_contexts' : n_contexts,
       'alpha' : alpha, 'alpha_rout' : alpha_rout,
       'sigma_in' : sigma_in, 'sigma_targ' : sigma_targ,
       'sigma_basal_context' : sigma_basal_context, 'sigma_apical_context' : sigma_apical_context, 'h' : h}

TIME = 1.

T = int(np.floor(TIME/dt))

nStepOutTraining = 1000

t_shut = 10

JMAX = 100.
nu_targ = 0.005
```

```python
## Training Rec

gamma = 10.
def f(x,gamma):
    return np.exp(x*gamma)/(np.exp(x*gamma)+1)
#f1 = @(x)(gamma*exp(gamma*x))./(exp(gamma*x)+1)-(gamma*exp(2*gamma*x))./(exp(gamma*x)+1).^2;

apicalFactorTrain = 1
apicalFactorTest = 0

nIterRec = 100

#eta = 5.
#eta_out = 0.1
#etaW = .0

eta = 10.
eta_out = 0.01
etaW = .0


def training_rec(print_err=True):
    
    print('Training...')
    
    for iter in range(nIterRec):
        
        ###### Online-Training
    
        #initialize simulation
        
        for cont_index in range(n_contexts):
        
            lttb.cont = lttb.cont*0
            lttb.cont[cont_index] = 1
            
            lttb.y_targ = lttb.y_targ_collection[cont_index]
            
    
            lttb.initialize(par)
    
            #run simulation
            dH = 0
    
            for t in range(lttb.T-2):
                
                lttb.step(apicalFactor = apicalFactorTrain)
    
                dH = dH*(1-dt/tau_m) + dt/tau_m*lttb.S_filt[:,t]
    
                DJ = np.outer(( lttb.S_apic_dist[:,t+1] - f(lttb.VapicRec[:,t],gamma) )*(1-lttb.S_apic_dist[:,t]) ,dH)
                lttb.J =  lttb.J + eta*DJ
    
                SR = lttb.S_filtRO[:,t+1]
                #SR = lttb.B_filt_rec[:,t+1]
                Y = lttb.Jout@SR
                DJRO = np.outer(lttb.y_targ[:,t+1] - Y,SR.T)
                lttb.Jout =  lttb.Jout + eta_out*DJRO
        
        ###### Test
        
        if iter%5==0:
            if print_err:
                print('t=' + str(iter))
            for cont_index in range(n_contexts):
    
                lttb.cont = lttb.cont*0
                lttb.cont[cont_index] = 1
                
                lttb.y_targ = lttb.y_targ_collection[cont_index]
    
                lttb.initialize(par)
    
                #run simulation
    
                for t in range(lttb.T-2):
                
                    lttb.step(apicalFactor = apicalFactorTest)

                SR = lttb.S_filtRO[:,1:-2]
                #SR = lttb.B_filt_rec[:,1:-2]
                Y = lttb.Jout@SR
                mse_rec_train = np.std(lttb.y_targ[:,1:-2] - Y)**2
                
                if print_err:
                    print(' ', mse_rec_train)
    
    print()
```

```python
def context_experiment(exp_type, t_sw, show_plots=False):
    
    print('Context experiment=%s, t_sw=%d' % (exp_type,t_sw))
    
    mses_onDiag = []
    mses_offDiag = []
    mses_before_onDiag = []
    mses_before_offDiag = []
    mses_after_onDiag = []
    mses_after_offDiag = []
    
    for cont_index in range(n_contexts):

        context = []

        lttb.cont = lttb.cont*0
        lttb.cont[cont_index] = 1

        lttb.y_targ = lttb.y_targ_collection[cont_index]

        lttb.initialize(par)

        #run simulation

        apicalFactor = 0

        for t in range(lttb.T-2):

            if t==t_sw:
                apicalFactor = 0
                #lttb.cont = lttb.cont*0.1
                if exp_type=='switch_context':
                    lttb.cont[0] = 1 - lttb.cont[0]
                    lttb.cont[1] = 1 - lttb.cont[1]
                elif exp_type=='turn_off_context':
                    lttb.cont[0] = 0
                    lttb.cont[1] = 0
            context.append(lttb.cont[cont_index])

            lttb.step(apicalFactor = apicalFactor)

        SR = lttb.S_filtRO[:,1:-2]
        #SR = lttb.B_filt_rec[:,1:-2]
        Y = lttb.Jout@SR

        if exp_type=='turn_off_context':
            mse_rec_train_before_onDiag = np.std(lttb.y_targ_collection[cont_index][:,1:t_sw+1] - Y[:,0:t_sw])**2
            mse_rec_train_before_offDiag = np.std(lttb.y_targ_collection[1-cont_index][:,1:t_sw+1] - Y[:,0:t_sw])**2
            mse_rec_train_after_onDiag = np.std(lttb.y_targ_collection[cont_index][:,t_sw+1:-2] - Y[:,t_sw:])**2
            mse_rec_train_after_offDiag = np.std(lttb.y_targ_collection[1-cont_index][:,t_sw+1:-2] - Y[:,t_sw:])**2
        elif exp_type=='switch_context':
            mse_rec_train_before_onDiag = np.std(lttb.y_targ_collection[cont_index][:,1:t_sw+1] - Y[:,0:t_sw])**2
            mse_rec_train_before_offDiag = np.std(lttb.y_targ_collection[1-cont_index][:,1:t_sw+1] - Y[:,0:t_sw])**2
            mse_rec_train_after_onDiag = np.std(lttb.y_targ_collection[1-cont_index][:,t_sw+1:-2] - Y[:,t_sw:])**2
            mse_rec_train_after_offDiag = np.std(lttb.y_targ_collection[cont_index][:,t_sw+1:-2] - Y[:,t_sw:])**2
        elif exp_type=='none':
            mse_rec_train_onDiag = np.std(lttb.y_targ_collection[cont_index][:,1:-2] - Y)**2
            mse_rec_train_offDiag = np.std(lttb.y_targ_collection[1-cont_index][:,1:-2] - Y)**2
        
        if exp_type=='turn_off_context' or exp_type=='switch_context':
            mses_before_onDiag.append(mse_rec_train_before_onDiag)
            mses_before_offDiag.append(mse_rec_train_before_offDiag)
            mses_after_onDiag.append(mse_rec_train_after_onDiag)
            mses_after_offDiag.append(mse_rec_train_after_offDiag)
            #print('  mse_before[%d]=%.3f\n  mse_after[%d]=%.3f' % (cont_index,mse_rec_train_before,cont_index,mse_rec_train_after))
        elif exp_type=='none':
            mses_onDiag.append(mse_rec_train_onDiag)
            mses_offDiag.append(mse_rec_train_offDiag)
            #print('  mse[%d]=%.3f' % (cont_index,mse_rec_train_onDiag))
        
    #mse, Y = lttb.train_ro(par,out_epochs = 1)

        if show_plots:

            plt.figure(figsize=(12, 4))
            plt.subplot(211)
            for i in range(len(Y)):
                plt.plot(Y[i].T, color='C'+str(i))
                if exp_type=='switch_context':
                    plt.plot(lttb.y_targ_collection[cont_index][i].T[0:t_sw], '--', color='C'+str(i))
                    plt.plot(range(t_sw,1000),lttb.y_targ_collection[1-cont_index][i].T[t_sw:], '--', color='C'+str(i))
                else:
                    plt.plot(lttb.y_targ[i].T, '--', color='C'+str(i))
            plt.plot(np.array(context), '--', lw=2, color=('black' if cont_index==0 else 'red'))
            if exp_type=='switch_context':
                plt.plot(1-np.array(context), '--', lw=2, color=('red' if cont_index==0 else 'black'))

            plt.xlabel('time(s)')
            plt.ylabel('$y_{targ}$ --- $y_{out}$')
            plt.subplot(212)
            plt.imshow(lttb.S_filtRO[0:20,:],aspect='auto')
            #plt.imshow(lttb.B_filt_rec[0:20,:],aspect='auto')

            #plt.subplot(313)
            #plt.imshow(lttb.S_wind[0:20,:],aspect='auto')
            
            if par['sigma_apical_context']>0 and par['sigma_basal_context']==0:
                sigma_context = par['sigma_apical_context']
                compartment = 'sopra'
            elif par['sigma_apical_context']==0 and par['sigma_basal_context']>0:
                sigma_context = par['sigma_basal_context']
                compartment = 'sotto'
            if exp_type=='switch_context':
                fig_title = "turn_off del contesto da %s [sigma=%.0f, contesto n. %d, mse=%.3f]" % (compartment,sigma_context,cont_index,mse_rec_train)
                fig_name = "turnoff_%s_sigma%.0f_contesto%d" % (compartment,sigma_context,cont_index)
            elif exp_type=='turn_off_context':
                fig_title = "switch del contesto da %s [sigma=%.0f, contesto n. %d]\nmse_before=%.3f, mse_after=%.3f" % (compartment,sigma_context,cont_index,mse_rec_train_before, mse_rec_train_after)
                fig_name = "switch_%s_sigma%.0f_contesto%d" % (compartment,sigma_context,cont_index)
            plt.suptitle(fig_title)
            plt.savefig('./figures/' + fig_name + '.pdf', transparent=False)

            plt.show()
            
    if exp_type=='turn_off_context' or exp_type=='switch_context':
        print('    onDiag: (<mse_before>,<mse_after>) = (%.3f,%.3f)' % (np.mean(mses_before_onDiag),np.mean(mses_after_onDiag)))
        print('    offDiag: (<mse_before>,<mse_after>) = (%.3f,%.3f)' % (np.mean(mses_before_offDiag),np.mean(mses_after_offDiag)))
    elif exp_type=='none':
        print('    onDiag: <mse>=%.3f' % (np.mean(mses_onDiag)))
        print('    offDiag: <mse>=%.3f' % (np.mean(mses_offDiag)))
    
    print()
```
